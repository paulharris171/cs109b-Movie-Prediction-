```python
#import libraries
%matplotlib inline
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import metrics
from sklearn.cross_validation import cross_val_score
import statsmodels.api as sm
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/paulharris171/cs109b-Movie-Prediction-/master/movie_dataframe.csv?token=AGPPqNosb_L40PCLX19Q_2mwtMoxEBsEks5ZAT4kwA%3D%3D")
```

```python
data_features = data[['director_name','budget','actor_1_name','content_rating','plot_keywords']]
data_y = data[['genre_adjusted']]
data_features = pd.get_dummies(data_features, columns=['content_rating'])
```

```python
data_features
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_features, data_y, test_size=0.33, random_state=42)
```

```python
vectorizer = CountVectorizer(analyzer = "word",   
                             tokenizer = None,    
                             preprocessor = None, 
                             stop_words = None,   
                             max_features = 5000) 
```

```python
train_data_features = vectorizer.fit_transform(data['plot_keywords'])
```

```python
train_data_features = train_data_features.toarray()
```

```python
print train_data_features.shape
#1153 rows and 3529 vocabulary words
```

```python
vocab = vectorizer.get_feature_names()
print vocab
```

```python
dist = np.sum(train_data_features, axis=0)
for tag, count in zip(vocab, dist):
    print count, tag
```

```python
#Try a random forest model 
forest = RandomForestClassifier(n_estimators = 100) 
forest = forest.fit(train_data_features, data['genre_adjusted'].values.ravel())
```

```python
#Use to predict genre on test set 
test_data_features = vectorizer.transform(X_test)
test_data_features = test_data_features.toarray()

# Use the random forest to make genre label predictions
result = forest.predict(test_data_features)
```
