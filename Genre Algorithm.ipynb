```python
genres = []
for title in titles:
    if len(title) == 1:
        genres.append(title[0])
    else:
        if 'Horror' in title:
            genres.append('Horror')
            continue
        elif 'Sport' in title:
            genres.append('Sport')
            continue
        elif 'War' in title:
            genres.append('War')
            continue
        elif 'Western' in title:
            genres.append('Western')
            continue
        elif 'Biography' in title: 
            genres.append('Biography')
            continue
        elif 'History' in title or 'Documentary' in title:
            genres.append('History')
            continue
        elif 'Animation' in title:
            genres.append('Animation')
            continue
        elif 'Family' in title:
            genres.append('Family')
            continue
        elif 'Music' in title or 'Musical' in title:
            genres.append('Musical')
            continue
        elif 'Romance' and 'Comedy' in title:
            genres.append('Rom-Com')
            continue
        elif 'Sci-Fi' in title:
            genres.append('Sci-Fi')
            continue
        elif 'Mystery' in title:
            genres.append('Mystery')
            continue
        elif 'Romance' and 'Action' in title:
            genres.append('Action-Romance')
            continue
        elif 'Comedy' and 'Action' in title:
            genres.append('Action-Comedy')
            continue
        elif 'Drama' and 'Action' in title:
            genres.append('Action-Drama')
            continue
        elif 'Thriller' and 'Action' in title:
            genres.append('Action-Thriller')
            continue
        genres.append(title[0])
```

```python
Counter(genres)
```

```python
data['genre_adjusted'] = genres
```
